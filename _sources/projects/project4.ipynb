# Проект IV

## Интерактивная карта на любую интересную тему


#### **Цель:**

**Создать интерактивную карту** на интересующую вас тему и опубликовать её.

#### **Требования к итоговому результату:**

- **Слои:**

  - минимум **4 разных тематических слоя**
  - минимум **3 разных способа изображения данных**:
    - картограмма (choropleth)
    - катодиаграмма (например, маркеры с кругами)
    - кластеризация маркеров
    - модель плотности (heatmap)
    - маркеры

- **Другие обязательные элементы:**

  - **Layer Control** (управление слоями)
  - минимум **2 дополнительных интерактивных инструмента** на выбор:
    - Mouse Position
    - Fullscreen
    - Mini Map
    - Search
    - или другие плагины Folium

- **Публикация:**  
  Карта должна быть размещена онлайн (рекомендуется `gh-pages` на GitHub, но можно использовать другие доступные способы).

- **README:**  
  К карте также прикладывается файл README, в котором вы даете ссылку на опубликованную карту, коротко в свободной форме описываете, о чем ваша работа и исходные данные с указанием источников

#### **Формат сдачи:**

- Jupyter Notebook с кодом + используемые данные
- Файл `README.md` с описанием
- Ссылка опубликованную карту
